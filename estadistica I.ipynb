```python
def estadisticos (datos):
    media= datos.mean()
    mediana= datos.median()
    desviacion= datos.std()
    minimo= datos.min()
    maximo=datos.max()
    return display( f"media:{media:.2f}, mediana:{mediana:.2f},desviacion: {desviacion:.2f},minimo: {minimo:.2f},maximo: {maximo:.2f}")
    
```
