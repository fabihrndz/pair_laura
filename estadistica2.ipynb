2. PIB, emisiones y población: ¿qué se relaciona con qué?
Variables:

gdp

co2-emissions

population

Calcula una matriz de correlación solo con estas tres variables.

```python
df_corr_reducido = df[['gdp', 'co2-emissions', 'population']].corr()
```

```python
plt.figure(figsize = (15, 15))

# creamos una matriz booleana de máscara
mask = np.triu(np.ones_like(df_corr_reducido, dtype=bool))  # np.triu para que ponga los datos una vez sin repetidos
                                                           # np.ones_like para que queden del mismo tamaño
# creamos un heatmap con la máscara
sns.heatmap(df_corr_reducido, 
            annot = True,                # para que aparezcan los numeros dentro de los cuadros
            fmt = ".2f",                 # dos decimales
            cmap = "mako",               # paleta de colores
            vmax = 1,                  
            vmin = -1, 
            mask = mask);
```
