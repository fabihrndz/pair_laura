```python
valores = [random.randint(1, 100) for _ in range(7)]
print(valores)
```

```python
indices = ["lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"]
```

```python
laserie = pd.Series(valores, index= indices)
print(laserie)
```

```python
dicc = {}

for x in zip(valores, indices):
    dicc.update(x)
```
