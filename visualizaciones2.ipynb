¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
df_c = df.groupby("continent")["country"].count().reset_index()
```

```python
sns.countplot(data=df, x= "continent", hue= "country");
```

```python
df_c
```

```python
sns.barplot(data= df_c, y= "continent", x= "country")
```

```python
sns.barplot(data= df_c, y= "continent", x= "country")
```

¿Cuáles son los 10 países con la mayor área forestal?

```python
df.head(2)
```

```python
df_a = df.groupby("country")["forested_area"].max().sort_values(ascending=False).head(10).reset_index()
df_a
```

```python
sns.barplot(data=df_a, y="country", x="forested_area", palette="viridis")
```

```python
sns.barplot(data=df_a, x="country", y="forested_area", hue="country")
plt.xticks(rotation = 90)
plt.title("Mayor area forestal por pais", fontsize= 14)
plt.xlabel("Pais", fontsize= 12)
plt.ylabel("Area forestal", fontsize= 12)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines["top"].set_visible(False)
;
```
